```python
# Importando as bibliotecas a serem usadas.
import numpy as np
import random as rd
```

```python
n = 0 # Contagem das leituras.
# Criação do vetor aleatório dos valores das pressões para a medição de 2 em 2 minutos, ou seja, 10 valores de pressão.
p = (np.random.uniform(1, 6, size=10))
print(p)
A = 0 # Contagem do alarme.
```

```python
# Criação do for para varer todo o vetor de pressão criado.
for pressao in p:
    if pressao >= 2 and pressao <= 5: # Verificar se a pressão está dentro do padrão.
     n = n+1 # Fazendo a contagem de leituras.
     A = A+0 # Fazendo a contagem dos alarmes.
     P = pressao # Chamando o valor da pressão.
     if A < 3: # Verificar se não atingiu o número máximo de alarmes.
      print(f'Leitura: {n}, Pressão: {P:.4f} bar, Alarmes ativados: {A}, Sistema: Operando')
     else: # Verificar se atingiu o número máximo de alarmes.
      print(f'Leitura: {n}, Pressão: {P:.4f} bar, Alarmes ativados: {A}, Sistema: Interrompido')
      break # Para parar o código.

    if pressao < 2 or pressao > 5: # Verificar se a pressão está fora do padrão.
      n = n+1 # Fazendo a contagem de leituras.
      A = A+1 # Fazendo a contagem dos alarmes.
      P = pressao # Chamando o valor da pressão.
      if A < 3: # Verificar se não atingiu o número máximo de alarmes.
       print(f'Leitura: {n}, Pressão: {P:.4f} bar, Alarmes ativados: {A}, Sistema: Operando')
      else: # Verificar se atingiu o número máximo de alarmes.
       print(f'Leitura: {n}, Pressão: {P:.4f} bar, Alarmes ativados: {A}, Sistema: Interrompido')
       break # Para parar o código.
```
